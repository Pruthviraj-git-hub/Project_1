**Winsoriazion**

- whatever the outlies less the lb fill with lb

- whatever the outlies less the ub fill with ub

- in numpy we have method **clip**

```python
pwag = visa_df['prevailing_wage']
step 1 : get the wage data values
step 2 : calcualte lb
step 3 : calcualte up
check the parameters and apply 
```

```python
pwage = visa_df['prevailing_wage']

Q1 = np.percentile(pwage,25)
Q2 = np.percentile(pwage,50)
Q3 = np.percentile(pwage,75)
iqr = Q3 - Q1

lb = Q1 - 1.5*iqr

up = Q3 + 1.5*iqr

wage_data_clip = np.clip(pwage,lb,ub)
```

```python
len(wage_data_clip)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1).hist(pwage)
plt.subplot(2,2,2).hist(wage_data_clip)
plt.subplot(2,2,3).boxplot(pwage,vert=False)
plt.subplot(2,2,4).boxplot(wage_data_clip,vert=False)
plt.show()
```
