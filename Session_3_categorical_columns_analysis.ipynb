```python
# step 1 : create a folder
folder_name = 'frequency_table'
os.makedirs(folder_name,exist_ok=True)

# step 2 : make a new path
directory = os.getcwd()  
new_dir = os.path.join(directory,folder_name)

#step-3:Create a DataFrame
for i in cat[1:]:
    keys = visa_df[i].value_counts().keys()
    values = visa_df[i].value_counts().values
    df = pd.DataFrame(zip(keys,values),
                      columns=['classes','frequency'])
    
    # step 4 : make a new path 
    file_name = f'{i}_df.csv'  # we are adding another file to the new path
    path = os.path.join(new_dir,file_name)
    
    # step-5 : save into dataframe
    df.to_csv(path,index=False)
```

```python
folder_name = 'bar_charts'
os.makedirs(folder_name,exist_ok=True)

directry = os.getcwd()
new_dir = os.path.join(directory,folder_name)

for i in cat[1:]:
    keys = visa_df[i].value_counts().keys()
    values = visa_df[i].value_counts().values
    plt.figure(figsize=(10,4))
    plt.bar(keys,values)
    plt.xlabel('classse')
    plt.ylabel('frequency')
    barchart_name = plt.title(f'{i}_barchart')
    
    file_name = f'{barchart_name}.jpg'
    path = os.path.join(new_dir,file_name)
    plt.savefig(path)
      
```

```python
folder_name = 'pie_chart'
os.makedirs(folder_name,exist_ok=True)

directry = os.getcwd()
new_dis = os.path.join(directry,folder_name)

cat = visa_df.select_dtypes(include='object').columns

for i in cat[1:]:
    keys = visa_df[i].value_counts().keys()
    values = visa_df[i].value_counts().values
    plt.pie(values,
            labels=keys,
            autopct='%1.1f%%')
    name = plt.title(f'{i}_piecahrt')
    plt.show()
    file_name =f'{name}.jpg'

    path = os.path.join(new_dis,file_name)
    plt.savefig(path)
```
